```python

import matplotlib.pyplot as plt
import networkx as nx
import random

# Function to create a random Directed Acyclic Graph (DAG)
def create_random_dag(num_nodes, num_edges):
    G = nx.DiGraph()
    for i in range(num_nodes):
        G.add_node(i)

    while len(G.edges) < num_edges:
        a, b = random.sample(range(num_nodes), 2)
        if a != b and not G.has_edge(b, a):
            G.add_edge(a, b)
            if nx.is_directed_acyclic_graph(G):
                continue
            else:
                G.remove_edge(a, b)
    return G

# Creating a random DAG with 10 nodes and 15 edges
num_nodes = 10
num_edges = 15
dag = create_random_dag(num_nodes, num_edges)

# Drawing the graph
plt.figure(figsize=(10, 8))
nx.draw_networkx(dag, node_color='skyblue', with_labels=True, node_size=700, arrowstyle='->', arrowsize=20)
plt.title("Random Directed Acyclic Graph (DAG)")
plt.show()
```
